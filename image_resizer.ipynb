```python
import os
import numpy as np
from PIL import Image
#Resizing images to 64x64
IMAGE_SIZE = 64
IMAGE_CHANNELS = 2
IMAGE_DIR = 'drive/MyDrive/input'

images_path = IMAGE_DIR
```

```python
#Here, we are using Pillow to resize all images to our desired size and appending them on a list as numpy array.
training_data = []
for filename in os.listdir(images_path):
    path = os.path.join(images_path, filename)
    image = Image.open(path).resize((IMAGE_SIZE, IMAGE_SIZE), Image.ANTIALIAS)
    training_data.append(np.asarray(image))
    
    
```

```python
#64x64 images are saved in a output file for later using as train data.
output_path = 'drive/MyDrive/Dataset/landscapeResized/resized'
if not os.path.exists(output_path):
    os.makedirs(output_path)
for i in range(len(training_data)):
    filename = os.path.join(output_path, f"resized-{i}.png")
    im = Image.fromarray(training_data[i])
    im.save(filename)
```
